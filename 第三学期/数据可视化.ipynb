```python
import h5py
import numpy as np
f = h5py.File('D:\library\Github\Innovative-practice\library\csv\第三学期\data_git_version.h5', 'r')
for key in f.keys():
    print(f[key].name)
#打印data关键字
print(f['data'].shape)
#(1488, 10000, 3) 1488个样本(每1h)，每个样本10000个点（100x100），每个点3个特征（sms, call, internet） 
print(f['idx'][0])
print(f['idx'][1])
#idx为每个样本的时间戳，第一个样本的时间戳为2013-11-01 01:00:00，第二个样本的时间戳为2013-11-01 02:00:00
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data0 = f['data'][0, :, 0]
data1 = f['data'][0, :, 1] 
data2 = f['data'][0, :, 2]
#将data转化成二维数据
data0 = data0.reshape(100, 100)
data1 = data1.reshape(100, 100)
data2 = data2.reshape(100, 100)
#将data转化成热力图
plt.figure( figsize=(10,30) )
plt.subplot(3,1,1)
sns.heatmap(data0)
plt.subplot(3,1,2)
sns.heatmap(data1)
plt.subplot(3,1,3)
sns.heatmap(data2)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure()
ax = fig.gca(projection='3d')
x = np.arange(0, 100, 1)
y = np.arange(0, 100, 1)
x, y = np.meshgrid(x, y)
data0 = f['data'][0, :, 2]
data0 = data0.reshape(100, 100)
Z = data0

surf = ax.plot_surface(x, y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('t0时刻internet热力图')
 
plt.show()
```
