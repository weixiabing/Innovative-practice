```python
from keras.models import Sequential,load_model
import h5py
import matplotlib.pyplot as plt
import numpy as np
import os
import cupy as cp
```

```python
#选择要测试kpi sms=0 call=1 internet=2
kpi=1
if kpi==0:
    kpi_name='sms'
elif kpi==1:
    kpi_name='call'
else:
    kpi_name='internet'
```

```python

f = h5py.File('/content/drive/Shareddrives/gsd/dataset_convlstm_3/testX.h5', 'r')

testX = f['testX']
testX = testX[:,:,:,:,kpi]
f1 = h5py.File('/content/drive/Shareddrives/gsd/dataset_convlstm_3/testY.h5', 'r')
testY = f1['testY']
testY = testY[:,:,:,kpi]
testX=cp.array(testX)
testY=cp.array(testY)
testX=cp.reshape(testX,(testX.shape[0], testX.shape[1], testX.shape[2],testX.shape[3],1))
testY=cp.reshape(testY,(testY.shape[0], testY.shape[1], testY.shape[2],1))


print(testX.shape)
print(testY.shape)
```

```python
model = load_model(os.path.join("/content/drive/Shareddrives/gsd/dataset_convlstm",f"Test_{kpi_name}" + ".h5"))
testPredict = model.predict(testX.get())
print(testPredict.shape)
```

```python
f2 = h5py.File(f'/content/drive/Shareddrives/gsd/dataset_convlstm/testpred_{kpi_name}.h5', 'w')
f2.create_dataset('data', data=testPredict)
f2.close()
```
