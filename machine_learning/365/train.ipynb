```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
import tensorflow as tf
import datetime
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 读取npy trainX和trainY
trainX = np.load('/content/drive/MyDrive/Colab Notebooks/机器学习/365/trainX.npy')
trainY = np.load('/content/drive/MyDrive/Colab Notebooks/机器学习/365/trainY.npy')
print(trainX.shape)
print(trainY.shape)
```

```python
from keras.layers import Dropout

model = Sequential([
    keras.layers.ConvLSTM2D(filters=32, kernel_size=(5, 5),
                   input_shape=(3,71, 73, 1),
                   padding='same', return_sequences=True),
    keras.layers.BatchNormalization(),
    keras.layers.ConvLSTM2D(filters=64, kernel_size=(3, 3),
                   padding='same', return_sequences=True),
    keras.layers.BatchNormalization(),
    keras.layers.ConvLSTM2D(filters=128, kernel_size=(1, 1),
                   padding='same', return_sequences=False),
    Dropout(0.2),
    keras.layers.Conv2D(filters=1, kernel_size=(3, 3),
               activation='sigmoid',
               padding='same')
])
model.compile(loss='binary_crossentropy', optimizer='adam')
model.summary()
```

```python
# 定义回调
early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", patience=10)
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor="val_loss", patience=5)
log_dir = "/content/drive/MyDrive/Colab Notebooks/机器学习/365/logs/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
# 设置训练参数
epochs = 50
batch_size = 2

# 拟合模型.
model.fit(
    trainX,
    trainY,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=0.2,
    callbacks=[early_stopping, reduce_lr,tensorboard_callback],
)
model.save('/content/drive/MyDrive/Colab Notebooks/机器学习/365/model.h5')
```
