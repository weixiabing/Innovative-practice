```python
import numpy as np
import h5py

```

```python
#读取第四学期\data\test_result.h5

f = h5py.File('./data/testlist.h5', 'r')
testY = f['testY']
testY = np.array(testY)
f.close()
f=h5py.File('./data/test_result.h5', 'r')
test_result = f['data']
test_result = np.array(test_result)
f.close()

```

```python
print(testY[:,1,1,0])
```

```python

#画图
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.plot(testY[:,12,1,0], label='testY')
plt.plot(test_result[:,12,1,0], label='test_result')
plt.legend()
plt.show()
```

```python
#计算误差testY和test_result的r2_score mae mape
r2,mae,mape=[],[],[]
from sklearn.metrics import r2_score,mean_absolute_error,mean_absolute_percentage_error
for i in range (20):
    for j in range (20):
        r2.append(r2_score(testY[:,i,j,0],test_result[:,i,j,0]))
        mae.append(mean_absolute_error(testY[:,i,j,0],test_result[:,i,j,0]))
        mape.append(mean_absolute_percentage_error(testY[:,i,j,0],test_result[:,i,j,0]))
print('r2_score:',np.mean(r2))
print('mae:',np.mean(mae))
print('mape:',np.mean(mape))

        
```
