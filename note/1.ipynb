```python
from matplotlib import pyplot as plt
#fig=plt.figure(figsize=(20,8),dpi=80)
x=range(2,26,2)
#数据在x轴位置，是一个可迭代对象
y=[15,13,14.5,17,20,25,26,26,24,22,18,15]
#y轴位置
plt.plot(x,y)
plt.xticks(range(2,26,2))
plt.show()
#plt.savefig('sig_size.svg')
```

```python
from matplotlib import pyplot as plt
import random
#fig=plt.figure(figsize=(20,8),dpi=80)
x=range(0,120)
#数据在x轴位置，是一个可迭代对象
y=[random.randint(20,35) for i in range(120)]
#y轴位置
plt.plot(x,y)

plt.show()
```
