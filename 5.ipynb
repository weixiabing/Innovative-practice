```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.core.frame import DataFrame
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
pf=pd.read_csv('https://raw.githubusercontent.com/weixiabing/Innovative-practice/main/library/csv/pm/beijing_final1.csv')

#对pf["Value"]进行三次指数平滑预测
#第一次指数平滑
se = SimpleExpSmoothing(np.asarray(pf["Value"])).fit()
#第二次指数平滑
se2 = SimpleExpSmoothing(se.fittedvalues).fit()
#第三次指数平滑
se3 = SimpleExpSmoothing(se2.fittedvalues).fit()
#第一次指数平滑预测
predict_sunspots = se.forecast(12).rename("第一次指数平滑预测")
#第二次指数平滑预测
predict_sunspots2 = se2.forecast(12).rename("第二次指数平滑预测")
#第三次指数平滑预测
predict_sunspots3 = se3.forecast(12).rename("第三次指数平滑预测")
#指数平滑预测结果
predict_sunspots = pd.concat([pf["Value"],predict_sunspots,predict_sunspots2,predict_sunspots3],axis=1)
predict_sunspots.plot(figsize=(12,8))
plt.show()



```
